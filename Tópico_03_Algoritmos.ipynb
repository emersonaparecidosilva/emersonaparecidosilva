```python
salário = float(input("Digite o salário para cálculo do imposto: "))
base = salário
imposto = 0
if base > 3000:
	imposto = imposto + ((base - 3000) * 0.35)
	base = 3000
if base > 1000:
	imposto = imposto + ((base - 1000) * 0.20)
print("Salário: R$%6.2f Imposto: R$ %6.2f" % (salário, imposto))
```

2. Escreva um programa para aprovar o empréstimo bancário para compra de uma casa. O programa deve perguntar o valor da casa a comprar, o
salário e a quantidade de anos a pagar. O valor da prestação mensal não pode ser superior a 30% do salário. Calcule o valor da prestação como sendo o valor da casa a comprar dividido pelo número de meses a pagar.

```python
valorCasa =  float(input("Qual o valor da casa? "))
salarioComprador =  float(input("Qual a sua renda? "))
anosFinanciamento =  float(input("Em quantos anos deseja pagar? "))
mesFinanciamento = anosFinanciamento * 12
limiteComprometimentoRenda = salarioComprador * 0.30
valorParcela = valorCasa / mesFinanciamento

if valorParcela <= limiteComprometimentoRenda:
  print("Aprovado")
  print("Sua parcela ficou em: R$%6.2f, dentro dos R$%6.2f disponíveis de sua renda" %(valorParcela,limiteComprometimentoRenda))
else:
  print("Reprovado")
  print("Sua renda possibilita uma parcela de: R$%6.2f e a parcela desta casa ficou em: R$%6.2f" %(limiteComprometimentoRenda,valorParcela))
```

1. Calcular a conta de um telefone celular da empresa Tchau. Os planos da empresa Tchau são bem interessantes e oferecem preços diferenciados de acordo com a quantidade de minutos usados por mês. Abaixo de 200 minutos, a empresa cobra R\$0,20 por minuto. Entre 200 e 400 minutos, o preço é de R$ 0,18. Acima de 400 minutos, o preço por minuto é de R\$0,15. O programa abaixo resolve esse problema.

```python
minutos = int(input("Digite a quantidade de minutos utilizada neste mês: "))
if minutos < 200:
  preco = 0.20
else :
  if minutos <= 400:
    preco = 0.18
  else :
    if minutos > 400:
      preco = 0.15
conta = minutos * preco
print("O valor a ser pago = R$%6.2f" % conta)
```

# **elif**

1. Escreva um programa que leia dois números e que pergunte qual operação você deseja realizar. Você deve poder calcular a soma (+), subtração (-), multiplicação (*) e divisão (/). Exiba o resultado da operação solicitada.

```python
x = int(input("Digite o valor de X: "))
y = int(input("Digite o valor de Y: "))
operacao = input("Digite a Operação: (+, -, *, /): ")
if operacao == "+":
  z = x + y
  print("O valor de Z = %d e a operação foi %s" %(z,operacao))
elif operacao == "-":
  z = x - y
  print("O valor de Z = %d e a operação foi %s" %(z,operacao))
elif operacao == "*":
  z = x * y
  print("O valor de Z = %d e a operação foi %s" %(z,operacao))
elif operacao == "/":
  z = x / y
  print("O valor de Z = %d e a operação foi %s" %(z,operacao))
else:
  print("Operação Inválida")
```
